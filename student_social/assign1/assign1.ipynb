```python
import pandas as pd

df = pd.read_csv("./assign1.csv")
```

```python
df.head()
```

```python
df = df[df['Relationship_Status'] != 'Complicated']
df['Relationship_Status'] = df['Relationship_Status'].map({'In Relationship':True,'Single':False})
```

```python
df['Most_Used_Platform'].value_counts()
```

```python
video = ("YouTube", "TikTok")
social = ("Instagram", "Facebook", "Twitter", "LinkedIn", "VKontakte", "Snapchat")
messenger = ("WhatsApp", "WeChat", "LINE", "KakaoTalk")
```

```python
platform_map = {}
for p in video: platform_map[p] = 'Video'
for p in social: platform_map[p] = 'Social'
for p in messenger: platform_map[p] = 'Messenger'
```

```python
df['Category'] = df['Most_Used_Platform'].map(platform_map)
```

```python
df = df.drop(columns=['Most_Used_Platform'])
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Category'])
```
